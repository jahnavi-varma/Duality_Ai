```python
import glob

label_files = glob.glob("/content/duality_dataset/train_2/train2/labels/*.txt")
unique_classes = set()

for file in label_files:
    with open(file, "r") as f:
        for line in f:
            cls_id = int(line.strip().split()[0])
            unique_classes.add(cls_id)

print("Total unique classes:", len(unique_classes))
print("Class IDs:", sorted(unique_classes))
```

```python
import os

all_names = set()
for f in label_files:
    base = os.path.basename(f)
    # remove ID prefix, keep descriptive part
    parts = "_".join(base.split("_")[1:]).replace(".txt", "")
    all_names.add(parts)

print("Class names from filenames:", all_names)
```

```python
data_yaml = """
train: /content/duality_dataset/train_2/train2/images
val: /content/duality_dataset/train_2/val2/images

nc: 7
names: ['vlight_uncluttered','vlight_cluttered_room','light_cluttered_room',
        'vcluttered_hallway','light_uncluttered','cluttered_room','dark_clutter']
"""

with open("/content/duality_dataset/data.yaml", "w") as f:
    f.write(data_yaml)

print("✅ Correct data.yaml created at /content/duality_dataset/data.yaml")
```

```python
import torch
print("CUDA available:", torch.cuda.is_available())
print("Device:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU only")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")   # medium model for debug
model.train(
    data="/content/duality_dataset/data.yaml",
    epochs=60,
    imgsz=640,
    batch=16,
    name="duality_baseline"
)
```

```python
results = model.val()
print(results)
```
